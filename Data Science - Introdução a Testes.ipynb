```python
import pandas as pd

tmdb = pd.read_csv('./tmdb_5000_movies.csv')
tmdb.head()
```

```python
tmdb.describe()
```

```python
import seaborn as sns

ax = sns.distplot(tmdb.vote_average)
ax.set(xlabel='Nota média', ylabel='Densidade')
ax.set_title('Média de botos em filmes no TMDB 5000')
```

```python
ax = sns.distplot(tmdb.vote_average, norm_hist=False, kde = False)
ax.set(xlabel='Nota média', ylabel='Densidade')
ax.set_title('Média de botos em filmes no TMDB 5000')
```

```python
ax = sns.boxplot(tmdb.vote_average)
ax.set(xlabel='Nota média')
ax.set_title('Distribuição de nota média dos filmes do TMDB 5000')
```

```python
tmdb.query('vote_average == 0')
```

```python
tmdb10Votos = tmdb.query('vote_count >= 10')
tmdb10Votos.describe()
```

```python
ax = sns.distplot(tmdb10Votos.vote_average)
ax.set(xlabel='Nota média', ylabel='Densidade')
ax.set_title('Média de botos em filmes no TMDB 5000 dentre os filmes com mais de 10 votos')
```

```python
ax = sns.boxplot(tmdb.vote_average)
ax.set(xlabel='Nota média')
ax.set_title('Distribuição de nota média dos filmes do TMDB 5000 dentre os filmes com mais de 10 votos')
```

```python
notas = pd.read_csv('./ratings.csv')
notas.head()
```

```python
notaMediaFilme = notas.groupby('movieId').mean()['rating']
notaMediaFilme.head()
```

```python
ax = sns.distplot(notaMediaFilme.values)
ax.set(xlabel='Nota média', ylabel='Densidade')
ax.set_title('Média de votos em filmes no Movielens 100k')
```

```python
qtdVotosFilme = notas.groupby('movieId').count()
filmes10Votos = qtdVotosFilme.query('rating >= 10').index
filmes10Votos.values
```

```python
notaFilme10Votos = notaMediaFilme.loc[filmes10Votos.values]
notaFilme10Votos.head()
```

```python
ax = sns.distplot(notaFilme10Votos.values)
ax.set(xlabel='Nota média', ylabel='Densidade')
ax.set_title('Média de votos em filmes no Movielens 100k com 10 ou mais votos')
```

```python
ax = sns.boxplot(notaFilme10Votos.values)
ax.set(xlabel='Nota média')
ax.set_title('Distribuição de nota média dos filmes no Movielens 100k com 10 ou mais votos')
```

```python
ax = sns.distplot(notaFilme10Votos.values, 
                  hist_kws = {'cumulative':True},
                  kde_kws = {'cumulative':True})
ax.set(xlabel='Nota média', ylabel='% acumulada de filmes')
ax.set_title('Média de votos em filmes no Movielens 100k com 10 ou mais votos')
```

```python
ax = sns.distplot(tmdb10Votos.vote_average,
                 hist_kws = {'cumulative':True},
                  kde_kws = {'cumulative':True})
ax.set(xlabel='Nota média', ylabel='% acumulada de filmes')
ax.set_title('Média de botos em filmes no TMDB 5000 dentre os filmes com mais de 10 votos')
```

```python
print("Média dos filmes com pelo menos 10 votos:", notaFilme10Votos.mean())
```

```python
notaFilme10Votos[0:5].mean()
```

```python
import matplotlib.pyplot as plt
import numpy as np

np.random.seed(75243)
temp = notaFilme10Votos.sample(frac=1)

medias = [temp[0:i].mean() for i in range (1, len(temp))]

plt.plot(medias)
```

```python
from statsmodels.stats.weightstats import zconfint

zconfint(notaFilme10Votos)
```

```python
from statsmodels.stats.weightstats import DescrStatsW
descr = DescrStatsW(notaFilme10Votos)
descr.tconfint_mean()
```
