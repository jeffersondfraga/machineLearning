```
import pandas as pd
filmes_uri = 'https://raw.githubusercontent.com/alura-cursos/machine-learning-algoritmos-nao-supervisionados/master/movies.csv'
filmes = pd.read_csv(filmes_uri)
filmes.columns = ['filme_id', 'titulo', 'generos']
filmes.head()
```

```
generos = filmes.generos.str.get_dummies()
dadosFilmes = pd.concat([filmes, generos], axis=1)
dadosFilmes.head()
```
